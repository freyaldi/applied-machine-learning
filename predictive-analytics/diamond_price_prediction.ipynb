```python
#required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
#load the dataset
url = 'https://raw.githubusercontent.com/tidyverse/ggplot2/master/data-raw/diamonds.csv'
diamonds = pd.read_csv(url)
diamonds
```

```python
diamonds.info()
```

```python
diamonds.describe()
```

```python
# Handle Missing Values
x = (diamonds.x == 0).sum()
y = (diamonds.y == 0).sum()
z = (diamonds.z == 0).sum()

print("Number of zero values at x: ", x)
print("Number of zero values at y: ", y)
print("Number of zero values at z: ", z)
```

```python
diamonds.loc[(diamonds['z']==0)]
```

```python
diamonds = diamonds.loc[(diamonds[['x','y','z']]!=0).all(axis=1)]
diamonds.shape
```

```python
diamonds.describe()
```

```python
#visualize data using boxplot to look the outlier
sns.boxplot(x=diamonds['carat'])
```

```python
#Handle Outliers with IQR method
Q1 = diamonds.quantile(0.25)
Q3 = diamonds.quantile(0.75)
IQR=Q3-Q1
diamonds=diamonds[~((diamonds<(Q1-1.5*IQR))|(diamonds>(Q3+1.5*IQR))).any(axis=1)]

diamonds.shape
```
